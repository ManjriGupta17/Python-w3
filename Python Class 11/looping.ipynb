```python
#initialization condition  increment/decrement
i=1
while i<=10:
    print(i)
    i=i+1
```

```python
#Print First N Natural Number
n=int(input("Enter number upto which you want to print"))
i=1
while i<=n:
    print(i)
    i=i+1
```

```python
#Print First N Natural Number in Reverse Order 10 to 1
i=10
while i>=1:
    print(i)
    i=i-1
```

```python
#Print First N Natural Number in Reverse Order [n to 1]
n=int(input("Enter a number: "))
while n>=1:
    print(n)
    n=n-1
```

```python
#Python Program to Find Sum of First n Natural Numbers [Video 17]
n=int(input("enter a number: "))
i=1
sum=0
while(i<=n):
    sum=sum+i
    i=i+1
print("Sum=",sum)
```

```python
#Python Program to Find Sum of Square of First n Natural Numbers
n=int(input("Enter number: "))
i=1
sum=0
while(i<=n):
    sum=sum+(i*i)
    i=i+1
print("Sum=",sum)
```

```python
#Python Program to Find Sum of cube of First n Natural Numbers
n=int(input("Enter number: "))
i=1
sum=0
while i<=n:
    sum=sum+(i*i*i)
    i=i+1
print("Sum",sum)
```
